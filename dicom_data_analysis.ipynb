```python
import pydicom as dicom
import matplotlib.pylab as plt
import pandas as pd
```

```python
# specify your image path
image_path = '/Users/moni/Downloads/ADNI4-DICOM-nano-10514/002_S_0413/Field_Mapping/2019-08-27_09_39_37.0/S868733/ADNI_002_S_0413_MR_Field_Mapping__br_raw_20190828115108376_21_S868733_I1221062.dcm'
ds = dicom.dcmread(image_path)

plt.imshow(ds.pixel_array)
```
