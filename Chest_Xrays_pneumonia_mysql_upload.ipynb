```python
import uuid 
import mysql.connector
import os
from PIL import Image
import io
import base64
```

```python
def insertBLOB(user_id, filename, img):
    print("Inserting BLOB into normal_img table")
    try:
        connection = mysql.connector.connect(host='localhost',
                                             database='xray_images',
                                             user='root',
                                             password='passwdummy#',auth_plugin='mysql_native_password')

        cursor = connection.cursor()
        sql_insert_blob_query = """ INSERT INTO normal_img
                          (userid, filename,IMAGE) VALUES (%s,%s,%s)"""

      


        # Convert data into tuple format
        insert_blob_tuple = (user_id,filename, img)
        result = cursor.execute(sql_insert_blob_query, insert_blob_tuple)
        connection.commit()
        print("Image and file inserted successfully as a BLOB into xray_img table", result)
        cursor.close()
        connection.close()
        print("MySQL connection is closed") 

    except mysql.connector.Error as error:
        print("Failed inserting BLOB data into MySQL table {}".format(error))


  
```

```python
#storing normal images without pneumonia

img_path="/Users/moni/Documents/normal_images/"


for image_to_load in os.listdir(img_path):
    fullpath = os.path.join(img_path, image_to_load)
    #im = Image.open(fullpath)
    with open(fullpath, 'rb') as f:
        image = f.read()
    encodestring = base64.b64encode(image)
    #image_bytes = io.BytesIO()
    insertBLOB(str(uuid.uuid1()),image_to_load,encodestring)

    print(image_to_load)
```
